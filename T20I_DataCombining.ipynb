```python
def InternationalT20(Year, Start, Final):
    deliveries = pd.read_excel('Ball-by-Ball_T20International_Final.xlsx')
    deliveries_year = deliveries[deliveries['Match_Id']>Start]
    deliveries_year = deliveries_year[deliveries_year['Match_Id']<Final]
    deliveries_year.set_index(np.arange(0,len(deliveries_year)), inplace = True)
    batsman_year = deliveries_year['Batsman'].unique()
    batsman_year.sort()
    Intl_year = pd.DataFrame(columns = ['Season', 'Player Name','Innings', 'Runs', 'Balls Faced', 'Outs', 'Not Outs', 'Total 4s', 'Total 6s', 'Batting Avg', 'Batting Strike Rate'])
    Intl_year['Player Name'] = np.arange(0, len(batsman_year))
    for i in range(0, len(batsman_year)):
        Intl_year['Player Name'][i] = batsman_year[i]
    Intl_year['Season'] = Year
    Intl_year[['Runs', 'Balls Faced', 'Outs', 'Total 4s', 'Total 6s']] = 0
    #DeathOver_year['Balls Faced'][115] = len(pd.DataFrame(deliveries_year.groupby(by = 'batsman'))[1][115])
    for index in range(0,len(Intl_year)):
        Intl_year['Balls Faced'][index] = len(pd.DataFrame(deliveries_year.groupby(by = 'Batsman'))[1][index])
        Intl_year['Runs'][index] = pd.DataFrame(deliveries_year.groupby(by = 'Batsman'))[1][index]['Batsman Runs'].sum()
        Intl_year['Outs'][index] = pd.DataFrame(deliveries_year.groupby(by = 'Batsman'))[1][index]['Player Dismissed'].count()
        Intl_year['Total 4s'][index] = len(pd.DataFrame(deliveries_year.groupby(by = 'Batsman'))[1][index][pd.DataFrame(deliveries_year.groupby(by = 'Batsman'))[1][index]['Batsman Runs'] == 4])
        Intl_year['Total 6s'][index] = len(pd.DataFrame(deliveries_year.groupby(by = 'Batsman'))[1][index][pd.DataFrame(deliveries_year.groupby(by = 'Batsman'))[1][index]['Batsman Runs'] == 6])
        Intl_year['Innings'][index] = len(pd.DataFrame(deliveries_year.groupby('Batsman'))[1][index]['Match_Id'].unique())
        Intl_year['Not Outs'][index] = Intl_year['Innings'][index] - Intl_year['Outs'][index]
        if Intl_year['Outs'][index] == 0:
            Intl_year['Batting Avg'][index] = "-"
        else:
            Intl_year['Batting Avg'][index] = round(Intl_year['Runs'][index]/ Intl_year['Outs'][index], 2)
        Intl_year['Batting Strike Rate'][index] = round((Intl_year['Runs'][index]/Intl_year['Balls Faced'][index])*100,2)
    return Intl_year
    #Intl_year.to_excel(Year+'DeathOverBatsman.xlsx')
```

```python
def InternationalT20Final(Year, Start, Final):
    deliveries = pd.read_excel('Ball-by-Ball_T20International_Final.xlsx')
    deliveries_year = deliveries[deliveries['Match_Id']>Start]
    deliveries_year = deliveries_year[deliveries_year['Match_Id']<Final]
    deliveries_year.set_index(np.arange(0,len(deliveries_year)), inplace = True)
    batsman_year = deliveries_year['Batsman'].unique()
    batsman_year.sort()
    bowlers_year = deliveries_year['Bowler'].unique()
    bowlers_year.sort()
    Players = pd.Series(np.concatenate((batsman_year, bowlers_year))).unique()
    Players = np.sort(Players)
    Intl_year = pd.DataFrame(columns = ['Season', 'Player Name','Innings', 'Runs', 'Balls Faced', 'Outs', 'Not Outs', 'Total 4s', 'Total 6s', 'Batting Avg', 'Batting Strike Rate','Bowling','Overs', 'Runs Conceded', 'Wickets', 'Bowling Avg', 'Bowling Econ', 'Bowling SR'])
    Intl_year['Player Name'] = np.arange(0, len(Players))
    for i in range(0, len(Players)):
        Intl_year['Player Name'][i] = Players[i]
    Intl_year['Season'] = Year
    #Intl_year[['Runs', 'Balls Faced', 'Outs', 'Total 4s', 'Total 6s']] = 0
    #DeathOver_year['Balls Faced'][115] = len(pd.DataFrame(deliveries_year.groupby(by = 'batsman'))[1][115])
    #return Intl_year['Player Name'][244]
    for index in range(0,len(Players)):
        for x in range(0,len(batsman_year)):
            if Intl_year['Player Name'][index] == batsman_year[x]:
                Intl_year['Balls Faced'][index] = len(pd.DataFrame(deliveries_year.groupby(by = 'Batsman'))[1][index])
                Intl_year['Runs'][index] = pd.DataFrame(deliveries_year.groupby(by = 'Batsman'))[1][index]['Batsman Runs'].sum()
                Intl_year['Outs'][index] = pd.DataFrame(deliveries_year.groupby(by = 'Batsman'))[1][index]['Player Dismissed'].count()
                Intl_year['Total 4s'][index] = len(pd.DataFrame(deliveries_year.groupby(by = 'Batsman'))[1][index][pd.DataFrame(deliveries_year.groupby(by = 'Batsman'))[1][index]['Batsman Runs'] == 4])
                Intl_year['Total 6s'][index] = len(pd.DataFrame(deliveries_year.groupby(by = 'Batsman'))[1][index][pd.DataFrame(deliveries_year.groupby(by = 'Batsman'))[1][index]['Batsman Runs'] == 6])
                Intl_year['Innings'][index] = len(pd.DataFrame(deliveries_year.groupby('Batsman'))[1][index]['Match_Id'].unique())
                Intl_year['Not Outs'][index] = Intl_year['Innings'][index] - Intl_year['Outs'][index]
                if Intl_year['Outs'][index] == 0:
                    Intl_year['Batting Avg'][index] = "-"
                else:
                    Intl_year['Batting Avg'][index] = round(Intl_year['Runs'][index]/ Intl_year['Outs'][index], 2)
                Intl_year['Batting Strike Rate'][index] = round((Intl_year['Runs'][index]/Intl_year['Balls Faced'][index])*100,2)
       
    for index in range(0, len(Players)):
        for x in range(0,len(bowlers_year)):
            if Intl_year['Player Name'][index] == bowlers_year[x]:  
                Intl_year['Overs'][index] = convertover(len(pd.DataFrame(deliveries_year.groupby(by = 'Bowler'))[1][index]))
                Intl_year['Runs Conceded'][index] = pd.DataFrame(deliveries_year.groupby(by = 'Bowler'))[1][index]['Batsman Runs'].sum() + pd.DataFrame(deliveries_year.groupby(by = 'Bowler'))[1][index]['Extra Runs'].sum()
                Intl_year['Wickets'][index] = pd.DataFrame(deliveries_year.groupby(by = 'Bowler'))[1][index]['Player Dismissed'].count()
                Intl_year['Bowling Avg'][index] = round(Intl_year['Runs Conceded'][index]/ Intl_year['Wickets'][index], 1)
                Intl_year['Bowling Econ'][index] = round(Intl_year['Runs Conceded'][index]/ Intl_year['Overs'][index], 1)
                Intl_year['Bowling SR'][index] = round(len(pd.DataFrame(deliveries_year.groupby(by = 'Bowler'))[1][index])/ Intl_year['Wickets'][index], 1)
    Intl_year.fillna('-')
    Intl_year.to_excel(str(Year)+'InternationalT20.xlsx')
```

```python
def convertover(balls):
    num = int(balls/6)
    num = float(str(num) + '.'+ str(balls%6))
    return num
```

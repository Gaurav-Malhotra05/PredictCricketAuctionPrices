```python
def Match_id(Year, start):
    Date = []
    team1 = []
    team2 = []
    venue = []
    match_number = []
    #for i in range(0, len(combined)):
    #    if combined[1][i] == 'date' and str(combined[2][i])[:4] == '2017':
    #        Date17.append(str(combined[2][i])[:10])
    for i in range(0, len(combined)):
        if combined[1][i] == 'date' and str(combined[2][i])[:4] == str(Year):
            Date.append(str(combined[2][i])[:10])
            if i < 209270:    
                math = 0
                for count in range(i-4, i+8):
                    if combined[1][count] == 'team' and combined[1][count+1] == 'team':
                        team1.append(combined[2][count])
                        team2.append(combined[2][count+1])
                    if combined[1][count] == 'venue':
                        venue.append(combined[2][count])
                    if combined[1][count] == 'match_number':
                        match_number.append(combined[2][count])
                    else:
                        math = math + 1
                        if math == 12:
                            match_number.append(0)
                        
                    
    df = pd.DataFrame(columns = ['Match ID','Date', 'Team', 'Team 2', 'Venue', 'Match Number'])
    df['Date'] = Date
    df['Team'] = team1
    df['Team 2'] = team2
    df['Venue'] = venue
    df['Match Number'] = match_number
    df['Match ID'] = np.arange(start, start+ len(Date))
    df.to_excel(str(Year)+'Match_id.xlsx')
```

```python
def DeathOver(Year, Start, Final):
    deliveries_year = deliveries[deliveries['match_id']>Start]
    deliveries_year = deliveries_year[deliveries_year['match_id']<Final]
    deliveries_year.set_index(np.arange(0,len(deliveries_year)), inplace = True)
    batsman_year = deliveries_year['batsman'].unique()
    batsman_year.sort()
    DeathOver_year = pd.DataFrame(columns = ['Season', 'Player Name','Innings', 'Runs', 'Balls Faced', 'Outs', 'Not Outs', 'Total 4s', 'Total 6s', 'Batting Avg', 'Batting Strike Rate'])
    DeathOver_year['Player Name'] = np.arange(0, len(batsman_year))
    for i in range(0, len(batsman_year)):
        DeathOver_year['Player Name'][i] = batsman_year[i]
    DeathOver_year['Season'] = Year
    DeathOver_year[['Runs', 'Balls Faced', 'Outs', 'Total 4s', 'Total 6s']] = 0
    #DeathOver_year['Balls Faced'][115] = len(pd.DataFrame(deliveries_year.groupby(by = 'batsman'))[1][115])
    for index in range(0,len(DeathOver_year)):
        DeathOver_year['Balls Faced'][index] = len(pd.DataFrame(deliveries_year.groupby(by = 'batsman'))[1][index])
        DeathOver_year['Runs'][index] = pd.DataFrame(deliveries_year.groupby(by = 'batsman'))[1][index]['batsman_runs'].sum()
        DeathOver_year['Outs'][index] = pd.DataFrame(deliveries_year.groupby(by = 'batsman'))[1][index]['player_dismissed'].count()
        DeathOver_year['Total 4s'][index] = len(pd.DataFrame(deliveries_year.groupby(by = 'batsman'))[1][index][pd.DataFrame(deliveries_year.groupby(by = 'batsman'))[1][index]['batsman_runs'] == 4])
        DeathOver_year['Total 6s'][index] = len(pd.DataFrame(deliveries_year.groupby(by = 'batsman'))[1][index][pd.DataFrame(deliveries_year.groupby(by = 'batsman'))[1][index]['batsman_runs'] == 6])
        DeathOver_year['Innings'][index] = len(pd.DataFrame(deliveries_year.groupby('batsman'))[1][index]['match_id'].unique())
        DeathOver_year['Not Outs'][index] = DeathOver_year['Innings'][index] - DeathOver_year['Outs'][index]
        if DeathOver_year['Outs'][index] == 0:
            DeathOver_year['Batting Avg'][index] = "-"
        else:
            DeathOver_year['Batting Avg'][index] = round(DeathOver_year['Runs'][index]/ DeathOver_year['Outs'][index], 2)
        DeathOver_year['Batting Strike Rate'][index] = round((DeathOver_year['Runs'][index]/DeathOver_year['Balls Faced'][index])*100,2)
    DeathOver_year.to_excel(Year+'DeathOverBatsman.xlsx')
```

```python
def PowerPlay(Year, Start, Final):
    deliveries = pd.read_csv('deliveries.csv')
    deliveries = deliveries[deliveries['over'] <= 6]
    deliveries_year = deliveries[deliveries['match_id']>Start]
    deliveries_year = deliveries_year[deliveries_year['match_id']<Final]
    deliveries_year.set_index(np.arange(0,len(deliveries_year)), inplace = True)
    batsman_year = deliveries_year['batsman'].unique()
    batsman_year.sort()
    PowerPlay_year = pd.DataFrame(columns = ['Season', 'Player Name','Innings', 'Runs', 'Balls Faced', 'Outs', 'Not Outs', 'Total 4s', 'Total 6s', 'Batting Avg', 'Batting Strike Rate'])
    PowerPlay_year['Player Name'] = np.arange(0, len(batsman_year))
    for i in range(0, len(batsman_year)):
        PowerPlay_year['Player Name'][i] = batsman_year[i]
    PowerPlay_year['Season'] = Year
    PowerPlay_year[['Runs', 'Balls Faced', 'Outs', 'Total 4s', 'Total 6s']] = 0
    for index in range(0,len(PowerPlay_year)):
        PowerPlay_year['Balls Faced'][index] = len(pd.DataFrame(deliveries_year.groupby(by = 'batsman'))[1][index])
        PowerPlay_year['Runs'][index] = pd.DataFrame(deliveries_year.groupby(by = 'batsman'))[1][index]['batsman_runs'].sum()
        PowerPlay_year['Outs'][index] = pd.DataFrame(deliveries_year.groupby(by = 'batsman'))[1][index]['player_dismissed'].count()
        PowerPlay_year['Total 4s'][index] = len(pd.DataFrame(deliveries_year.groupby(by = 'batsman'))[1][index][pd.DataFrame(deliveries_year.groupby(by = 'batsman'))[1][index]['batsman_runs'] == 4])
        PowerPlay_year['Total 6s'][index] = len(pd.DataFrame(deliveries_year.groupby(by = 'batsman'))[1][index][pd.DataFrame(deliveries_year.groupby(by = 'batsman'))[1][index]['batsman_runs'] == 6])
        PowerPlay_year['Innings'][index] = len(pd.DataFrame(deliveries_year.groupby('batsman'))[1][index]['match_id'].unique())
        PowerPlay_year['Not Outs'][index] = PowerPlay_year['Innings'][index] - PowerPlay_year['Outs'][index]
        if PowerPlay_year['Outs'][index] == 0:
            PowerPlay_year['Batting Avg'][index] = "-"
        else:
            PowerPlay_year['Batting Avg'][index] = round(PowerPlay_year['Runs'][index]/ PowerPlay_year['Outs'][index], 2)
        PowerPlay_year['Batting Strike Rate'][index] = round((PowerPlay_year['Runs'][index]/PowerPlay_year['Balls Faced'][index])*100,2)
    PowerPlay_year.to_excel(Year+'PowerPlayBatsman.xlsx')
```

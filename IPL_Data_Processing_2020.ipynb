```python
def OutputYear(Year):
    batting = pd.read_excel(Year+' Batting.xlsx', sheet_name = 'Most Runs', header = 0)
    bowling = pd.read_excel(Year+' Bowling.xlsx', sheet_name = 'Most Wickets', header = 0)
    Batsman = np.array(batting['PLAYER'])
    Bowlers = np.array(bowling['PLAYER'])
    Players = pd.Series(np.concatenate((Batsman, Bowlers))).unique()
    df = pd.DataFrame(Players).sort_values(by = 0, ascending = True)
    df.to_excel(Year+'FinalOutput.xlsx', sheet_name = 'Output')
    output = df
    for count in range(1,len(output2013.columns)):
        output[output2013.columns[count]] = None
    output['Bowling Innings'] = None
    batting = batting[['PLAYER', 'Mat', 'Inns','NO', 'Runs','Avg', 'HS', 'SR', '4s', '6s', '50', '100']]
    for i in range(0,len(Players)):
        for x in range(0,len(Batsman)):
            if df[0][i] == batting['PLAYER'][x]:
                output['Matches'][i] = batting['Mat'][x]
                output['Innings'][i] = batting['Inns'][x]
                output['Not Outs'][i] = batting['NO'][x]
                output['Runs'][i] = batting['Runs'][x]
                output['Batting Avg'][i] = batting['Avg'][x]
                output['Highest Score'][i] = batting['HS'][x]
                output['Batting Strike Rate'][i] = batting['SR'][x]
                output['Total 4s'][i] = batting['4s'][x]
                output['Total 6s'][i] = batting['6s'][x]
                output['Total 50s'][i] = batting['50'][x]
                output['Total 100s'][i] = batting['100'][x]
    bowling = bowling[['PLAYER', 'Mat', 'Inns','Ov', 'Runs', 'Wkts', 'Avg', 'Econ', 'SR', '4w', '5w']]
    for i in range(0,len(Players)):
        for x in range(0,len(Bowlers)):
            if df[0][i] == bowling['PLAYER'][x]:
                output['Matches'][i] = bowling['Mat'][x]
                output['Bowling Innings'][i] = bowling['Inns'][x]
                output['Overs'][i] = bowling['Ov'][x]
                output['Runs Conceded'][i] = bowling['Runs'][x]
                output['Bowling Average'][i] = bowling['Avg'][x]
                output['Wickets Taken'][i] = bowling['Wkts'][x]
                output['Econ Rate'][i] = bowling['Econ'][x]
                output['Bowling Strike Rate'][i] = bowling['SR'][x]
                output['4w'][i] = bowling['4w'][x]
                output['5w'][i] = bowling['5w'][x]
    output.fillna('-', inplace=True)
    output.to_excel(Year+'FinalOutput.xlsx', sheet_name= 'FOutput', index = False)
```

```python
for year in ['2013','2014','2015','2016','2017','2018','2019']:
    OutputYear(str(year))
```

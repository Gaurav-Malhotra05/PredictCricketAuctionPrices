```python
def MasterSheet(Year):
    IPL = pd.read_excel('~/Desktop/Gaurav IPL Project/Final Output/Combined/IPL/'+str(Year)+'FinalOutput.xlsx')
    T20I = pd.read_excel('~/Desktop/Gaurav IPL Project/Final Output/Combined/T20I/'+str(Year)+'T20I_Combined_Table.xlsx')
    BLDeath = pd.read_excel('~/Desktop/Gaurav IPL Project/Final Output/Combined/BowlingDeathOvers/'+str(Year)+'Bowling_Death_Overs.xlsx')
    BLPower = pd.read_excel('~/Desktop/Gaurav IPL Project/Final Output/Combined/BowlingPowerPlay/'+str(Year)+'Bowling_Powerplay_Overs.xlsx')
    BAPower = pd.read_excel('~/Desktop/Gaurav IPL Project/Final Output/Combined/PowerPlayBatsman/'+str(Year)+'PowerPlayBatsman.xlsx')
    BADeath = pd.read_excel('~/Desktop/Gaurav IPL Project/Final Output/Combined/DeathOverBatsman/'+str(Year)+'DeathOverBatsman.xlsx')
    IPLName = pd.read_excel('~/Desktop/Gaurav IPL Project/Final Output/Combined/IPLNew/'+str(Year)+'.xlsx')
    BADeath.columns = BADeath.columns + ' Death'
    BAPower.columns = BAPower.columns+ ' PowerPlay'
    T20I.columns = T20I.columns + ' T20I'
    Master = pd.DataFrame(columns = ['Season', 'Player Name', 'DOB', 'Age',
       'Salary(Rupees Crore)', 'Role', 'Country', 'Intl./Dom.', 'Matches', 'Innings', 'Not Outs',
       'Runs', 'Batting Avg', 'Batting Strike Rate', 'Highest Score',
       'Total 4s', 'Total 6s', 'Total 50s', 'Total 100s','Bowling Innings', 'Overs',
       'Runs Conceded', 'Wickets Taken', 'Bowling Average', 'Econ Rate',
       'Bowling Strike Rate', '4w', '5w', 'DeathOver','Batting Avg Death',
       'Batting Strike Rate Death', 'Bowling Average in Death Overs',
       'Bowling SR in Death Overs', 'Bowling Econ. Rate in Death Overs', 'PowerPlay', 'Batting Avg PowerPlay',
       'Batting Strike Rate PowerPlay', 'Bowling Average in Powerplay Overs',
       'Bowling SR in Powerplay Overs',
       'Bowling Econ. Rate in Powerplay Overs', 'T20International', 'Batting Avg T20I', 'Batting Strike Rate T20I','Bowling Average T20I', 'Bowling SR T20I',
       'Bowling Econ. Rate T20I'])
    Master['Player Name'] = IPLName['Player Name']
    Master['Season'] = Year
    for i in ['DOB', 'Age',
       'Salary(Rupees Crore)', 'Role', 'Country', 'Intl./Dom.', 'Matches', 'Innings', 'Not Outs',
       'Runs', 'Batting Avg', 'Batting Strike Rate', 'Highest Score',
       'Total 4s', 'Total 6s', 'Total 50s', 'Total 100s','Bowling Innings', 'Overs',
       'Runs Conceded', 'Wickets Taken', 'Bowling Average', 'Econ Rate',
       'Bowling Strike Rate', '4w', '5w']:
        if i == 'DOB' and Year == 2015:
            Master[i] = IPL['DOB ']
        else:
            Master[i] = IPL[i]
            
    for count in range(0, len(Master)):
        for i in range(0, len(BADeath)):
            if BADeath['Player Name Death'][i] == Master['Player Name'][count]:
                #for blah in ['Batting Avg Death','Batting Strike Rate Death']:
                Master['Batting Avg Death'][count] = BADeath['Batting Avg Death'][i] 
                Master['Batting Strike Rate Death'][count] = BADeath['Batting Strike Rate Death'][i]
    
    for count in range(0, len(Master)):
        for i in range(0, len(BAPower)):
            if BAPower['Player Name PowerPlay'][i] == Master['Player Name'][count]:
                #for blah in ['Batting Avg Death','Batting Strike Rate Death']:
                Master['Batting Avg PowerPlay'][count] = BAPower['Batting Avg PowerPlay'][i] 
                Master['Batting Strike Rate PowerPlay'][count] = BAPower['Batting Strike Rate PowerPlay'][i] 
    
    for count in range(0, len(Master)):
        for i in range(0, len(BLPower)):
            if BLPower['Bowler Name'][i] == Master['Player Name'][count]:
                #'Bowling Average in Powerplay Overs','Bowling SR in Powerplay Overs','Bowling Econ. Rate in Powerplay Overs'
                Master['Bowling Average in Powerplay Overs'][count] = BLPower['Bowling Average in Powerplay Overs'][i] 
                Master['Bowling SR in Powerplay Overs'][count] = BLPower['Bowling SR in Powerplay Overs'][i] 
                Master['Bowling Econ. Rate in Powerplay Overs'][count] = BLPower['Bowling Econ. Rate in Powerplay Overs'][i]
                
    for count in range(0, len(Master)):
        for i in range(0, len(BLDeath)):
            if BLDeath['Bowler Name'][i] == Master['Player Name'][count]:
                #'Bowling Average in Powerplay Overs','Bowling SR in Powerplay Overs','Bowling Econ. Rate in Powerplay Overs'
                Master['Bowling Average in Death Overs'][count] = BLDeath['Bowling Average in Death Overs'][i] 
                Master['Bowling SR in Death Overs'][count] = BLDeath['Bowling SR in Death Overs'][i] 
                Master['Bowling Econ. Rate in Death Overs'][count] = BLDeath['Bowling Econ. Rate in Death Overs'][i]
    
    for count in range(0, len(Master)):
        for i in range(0, len(T20I)):
            if T20I['Player Name T20I'][i] == Master['Player Name'][count]:
                #'Bowling Average in Powerplay Overs','Bowling SR in Powerplay Overs','Bowling Econ. Rate in Powerplay Overs'
                Master['Batting Avg T20I'][count] = T20I['Batting Avg T20I'][i] 
                Master['Batting Strike Rate T20I'][count] = T20I['Batting Strike Rate T20I'][i] 
                Master['Bowling Average T20I'][count] = T20I['Bowling Average T20I'][i] 
                Master['Bowling SR T20I'][count] = T20I['Bowling SR T20I'][i] 
                Master['Bowling Econ. Rate T20I'][count] = T20I['Bowling Econ. Rate T20I'][i] 
                
    Master.fillna('-', inplace = True)
    Master.to_excel(str(Year)+'MasterSheetEdited.xlsx')
```
